```python
# ==============================================================================
# PROJET : PR√âDICTION MALADIE CARDIAQUE - Heart Disease Statlog
# ==============================================================================

# 1. IMPORTATION DES BIBLIOTH√àQUES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.impute import SimpleImputer  # Ajout de l'import ici
from scipy import stats
import os
import warnings
import kagglehub  # Pour le t√©l√©chargement du dataset

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid", palette="husl")
plt.rcParams["figure.figsize"] = (12, 8)
print("1. ‚úÖ Biblioth√®ques import√©es")

# 2. CHARGEMENT DU DATASET
print("2. üìä Chargement du dataset")
colonnes_attendues = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]

# T√©l√©chargement du dataset via KaggleHub
try:
    download_path = kagglehub.dataset_download("ritwikb3/heart-disease-statlog")
    print(f"   ‚úÖ Dataset t√©l√©charg√© dans : {download_path}")
except Exception as e:
    print(f"   ‚ùå Erreur lors du t√©l√©chargement du dataset via KaggleHub: {e}")
    download_path = None

df = None  # Initialisation de df
if download_path:
    csv_filename = "Heart_disease_statlog.csv"
    file_path = None
    # Recherche du fichier CSV dans le r√©pertoire t√©l√©charg√©
    for root, _, files in os.walk(download_path):
        if csv_filename in files:
            file_path = os.path.join(root, csv_filename)
            break
    if file_path:
        try:
            df = pd.read_csv(file_path)
            print(f"   ‚úÖ Succ√®s : {df.shape[0]} patients, {df.shape[1]} variables")
            if list(df.columns) != colonnes_attendues:
                print("   ‚ö†Ô∏è Colonnes inattendues. V√©rifie le dataset.")
            print(f"   ‚Ä¢ √Çge (min/max) : {df['age'].min()}-{df['age'].max()} ans")
            print(f"   ‚Ä¢ Cholest√©rol moyen : {df['chol'].mean():.0f} mg/dL")
            print(f"   ‚Ä¢ Tension art√©rielle moyenne : {df['trestbps'].mean():.0f} mmHg")
            print(f"   ‚Ä¢ Patients atteints : {df['target'].sum()} ({df['target'].mean()*100:.1f}%)")
        except Exception as e:
            print(f"   ‚ùå Erreur lors du chargement du dataset depuis '{file_path}': {e}")
    else:
        print(f"   ‚ùå Fichier '{csv_filename}' introuvable dans le r√©pertoire t√©l√©charg√©: {download_path}. V√©rifie le chemin ou le nom du fichier.")
else:
    print("   ‚ö†Ô∏è Le dataset n'a pas pu √™tre t√©l√©charg√©, impossible de proc√©der au chargement.")

# 3. SIMULATION DE DONN√âES "SALES" (NaN)
if df is not None:
    print("\n3. ‚ö†Ô∏è Simulation de 5% de valeurs manquantes (NaN) - sans toucher √† 'target'")
    np.random.seed(42)  # Pour la reproductibilit√©
    features_columns = [col for col in df.columns if col != 'target']  # Exclut 'target'
    df_dirty = df.copy()  # Copie du DataFrame original
    # Introduction de 5% de NaN dans les colonnes s√©lectionn√©es
    for col in features_columns:
        df_dirty.loc[df_dirty.sample(frac=0.05, random_state=42).index, col] = np.nan
    # Affichage du nombre total de valeurs manquantes g√©n√©r√©es
    total_nan = df_dirty.isnull().sum().sum()
    print(f"   ‚Ä¢ Nombre total de valeurs manquantes g√©n√©r√©es : {total_nan}")
    print("   ‚Ä¢ R√©partition des NaN par colonne :")
    print(df_dirty.isnull().sum())
else:
    print("   ‚ùå Le DataFrame 'df' n'a pas √©t√© charg√©. La simulation de valeurs manquantes ne peut pas √™tre effectu√©e.")

# 4. NETTOYAGE ET PR√âPARATION (Data Wrangling)
if df is not None:
    print("\n4. Nettoyage des donn√©es...")
    # S√©paration des Features (X) et de la Target (y) AVANT le nettoyage
    X, y = df_dirty.drop('target', axis=1), df_dirty['target']
    # Imputation des NaN par la moyenne
    X_clean = pd.DataFrame(
        SimpleImputer(strategy='mean').fit_transform(X),
        columns=X.columns  # Conservation des noms de colonnes
    )
    # V√©rification des NaN restants et statistiques post-imputation
    nan_restants = X_clean.isnull().sum().sum()
    print(f"   ‚úÖ Imputation termin√©e (NaN restants : {nan_restants}).")
    print("   üìä Statistiques apr√®s imputation (moyenne et √©cart-type) :")
    print(X_clean.describe().loc[['mean', 'std']].round(2))
else:
    print("   ‚ùå Le DataFrame 'df' n'a pas √©t√© charg√©. Le nettoyage et la pr√©paration ne peuvent pas √™tre effectu√©s.")

# 5. ANALYSE EXPLORATOIRE DES DONN√âES (EDA)
print("\n5. üîç Analyse Exploratoire des Donn√©es (EDA) - Version Finale Compl√®te")

# --- Fonctions utilitaires pour les graphiques ---
def plot_distribution(data, feature, hue, title, xlabel, ylabel, plot_type='kde'):
    """Trace une distribution (KDE ou histogramme) d'une feature selon une cible."""
    plt.figure(figsize=(12, 6))
    if plot_type == 'kde':
        sns.kdeplot(data=data, x=feature, hue=hue, fill=True, common_norm=False, palette='husl', alpha=0.5)
    else:
        sns.histplot(data=data, x=feature, hue=hue, kde=True, palette='husl', alpha=0.5)
    plt.title(title, fontsize=14, pad=20)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.tight_layout()
    plt.show()

def plot_categorical(data, feature, hue, title, xlabel, ylabel, types_dict=None):
    """Trace la r√©partition d'une variable cat√©gorielle selon une cible."""
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(data=data, x=feature, hue=hue, palette='Set2')
    plt.title(title, fontsize=14, pad=20)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    if types_dict:
        plt.xticks(ticks=range(len(types_dict)), labels=list(types_dict.values()), fontsize=10)
    plt.legend(title="Maladie cardiaque", labels=["Sain", "Malade"], fontsize=10)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 0.5, f"{int(height)}", ha="center", fontsize=10)
    plt.tight_layout()
    plt.show()

def plot_boxplot(data, feature, hue, title, xlabel, ylabel):
    """Trace un boxplot d'une feature selon une cible."""
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x=hue, y=feature, palette='pastel')
    plt.title(title, fontsize=14, pad=20)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.tight_layout()
    plt.show()

def plot_correlation_heatmap(data, target_column=None, title="Matrice de Corr√©lation"):
    """Trace une heatmap de corr√©lation."""
    plt.figure(figsize=(12, 8))
    if target_column:
        corr_matrix = data.corr()[target_column].sort_values(ascending=False).drop(target_column).to_frame()
    else:
        corr_matrix = data.corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1, annot_kws={"size": 10}, linewidths=0.5)
    plt.title(title, fontsize=14, pad=20)
    plt.xticks(fontsize=10, rotation=45)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

# --- A. Statistiques Descriptives ---
print("\nüìä Statistiques descriptives des variables cl√©s :")
display(X_clean[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']].describe().T.round(2))

# --- B. Distributions des Variables Cl√©s ---
print("\nüìà Distributions des variables cl√©s selon la maladie cardiaque :")

# B.1 Distribution de l'√¢ge
plot_distribution(
    data=df,
    feature='age',
    hue='target',
    title="Distribution de l'√¢ge selon la maladie cardiaque (0=Sain, 1=Malade)",
    xlabel="√Çge (ans)",
    ylabel="Densit√©"
)

# B.2 Distribution du cholest√©rol
plot_distribution(
    data=df,
    feature='chol',
    hue='target',
    title="Distribution du cholest√©rol selon la maladie cardiaque",
    xlabel="Cholest√©rol (mg/dL)",
    ylabel="Densit√©",
    plot_type='hist'
)

# --- C. Variables Cat√©gorielles ---
print("\nüè∑Ô∏è R√©partition des variables cat√©gorielles selon la maladie cardiaque :")

# C.1 Types de douleur thoracique (cp)
douleur_types = {0: "Typique angine", 1: "Atypique angine", 2: "Non-anginal", 3: "Asymptomatique"}
plot_categorical(
    data=df,
    feature='cp',
    hue='target',
    title="R√©partition des types de douleur thoracique selon la maladie cardiaque",
    xlabel="Type de douleur thoracique",
    ylabel="Nombre de patients",
    types_dict=douleur_types
)

# C.2 Sexe des patients
plot_categorical(
    data=df,
    feature='sex',
    hue='target',
    title="R√©partition des patients selon le sexe et la maladie cardiaque",
    xlabel="Sexe (0=Femme, 1=Homme)",
    ylabel="Nombre de patients"
)

# --- D. Corr√©lations avec la Cible ---
print("\nüîó Corr√©lations des variables avec la maladie cardiaque :")
X_with_target = X_clean.copy()
X_with_target['target'] = y
plot_correlation_heatmap(
    data=X_with_target,
    target_column='target',
    title="Corr√©lation des variables avec la maladie cardiaque (target)"
)

# --- E. Matrice de Corr√©lation entre Variables Cl√©s ---
print("\nüîó Matrice de corr√©lation entre les variables cl√©s :")
variables_interet = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
plot_correlation_heatmap(
    data=X_with_target[variables_interet],
    title="Matrice de corr√©lation entre les variables cl√©s"
)

# --- F. Comparaisons des Variables Cliniques ---
print("\nüìä Comparaisons des variables cliniques selon la maladie cardiaque :")

# F.1 Fr√©quence cardiaque maximale (thalach)
plot_boxplot(
    data=df,
    feature='thalach',
    hue='target',
    title="Fr√©quence cardiaque maximale (thalach) selon la maladie cardiaque",
    xlabel="Maladie cardiaque (0=Sain, 1=Malade)",
    ylabel="Fr√©quence cardiaque maximale (bpm)"
)

# F.2 D√©pression ST (oldpeak)
plot_boxplot(
    data=df,
    feature='oldpeak',
    hue='target',
    title="D√©pression ST (oldpeak) selon la maladie cardiaque",
    xlabel="Maladie cardiaque (0=Sain, 1=Malade)",
    ylabel="D√©pression ST (oldpeak)"
)

# --- G. Analyse des Outliers ---
print("\n‚ö†Ô∏è Analyse des outliers pour les variables cl√©s :")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
sns.boxplot(data=df, x='target', y='chol', ax=axes[0, 0], palette='pastel')
axes[0, 0].set_title("Niveau de cholest√©rol (chol) selon la maladie", fontsize=12)
axes[0, 0].set_xlabel("Maladie cardiaque (0=Sain, 1=Malade)", fontsize=10)
axes[0, 0].set_ylabel("Cholest√©rol (mg/dL)", fontsize=10)

sns.boxplot(data=df, x='target', y='trestbps', ax=axes[0, 1], palette='pastel')
axes[0, 1].set_title("Tension art√©rielle (trestbps) selon la maladie", fontsize=12)
axes[0, 1].set_xlabel("Maladie cardiaque (0=Sain, 1=Malade)", fontsize=10)
axes[0, 1].set_ylabel("Tension art√©rielle (mmHg)", fontsize=10)

sns.boxplot(data=df, x='target', y='thalach', ax=axes[1, 0], palette='pastel')
axes[1, 0].set_title("Fr√©quence cardiaque maximale (thalach) selon la maladie", fontsize=12)
axes[1, 0].set_xlabel("Maladie cardiaque (0=Sain, 1=Malade)", fontsize=10)
axes[1, 0].set_ylabel("Fr√©quence cardiaque maximale (bpm)", fontsize=10)

sns.boxplot(data=df, x='target', y='oldpeak', ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title("D√©pression ST (oldpeak) selon la maladie", fontsize=12)
axes[1, 1].set_xlabel("Maladie cardiaque (0=Sain, 1=Malade)", fontsize=10)
axes[1, 1].set_ylabel("D√©pression ST (oldpeak)", fontsize=10)

plt.tight_layout()
plt.show()

# --- H. Interactions entre Variables Cl√©s ---
print("\nüîÑ Interactions entre les variables cl√©s :")
g = sns.PairGrid(X_with_target[variables_interet + ['target']], hue='target', palette='husl', corner=True)
g = g.map_upper(sns.scatterplot, alpha=0.6, edgecolor='w')
g = g.map_lower(sns.kdeplot, levels=4, color='gray', alpha=0.4)
g = g.map_diag(sns.histplot, kde=True, alpha=0.7)
g = g.add_legend(title="Maladie cardiaque", labels=["Sain", "Malade"], bbox_to_anchor=(0.6, 0.8))
g.fig.suptitle("Interactions entre les variables cl√©s selon la maladie cardiaque", y=1.05, fontsize=14)
for ax in g.axes.flat:
    if ax:
        ax.set_xlabel(ax.get_xlabel(), fontsize=10)
        ax.set_ylabel(ax.get_ylabel(), fontsize=10)
plt.tight_layout()
plt.show()

# --- I. Relations Lin√©aires avec la Cible ---
print("\nüìâ Relations lin√©aires entre les variables cl√©s et la maladie cardiaque :")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
for i, variable in enumerate(variables_interet):
    sns.regplot(data=X_with_target, x=variable, y='target', ax=axes[i],
                scatter_kws={'alpha':0.4, 'edgecolor':'w'}, line_kws={'color':'red'})
    axes[i].set_title(f"Relation entre {variable} et la maladie", fontsize=12)
    axes[i].set_xlabel(variable, fontsize=10)
    axes[i].set_ylabel("Maladie cardiaque (0/1)", fontsize=10)
    axes[i].grid(True, linestyle='--', alpha=0.6)
if len(variables_interet) < 6:
    axes[-1].axis('off')
plt.tight_layout()
plt.show()

# 6. S√âPARATION DES DONN√âES (Train / Test Split)
X_train, X_test, y_train, y_test = train_test_split(X_clean, y, test_size=0.2, random_state=42, stratify=y)

# Normalisation des donn√©es
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 7. MOD√âLISATION (Gradient Boosting - XGBoost)
# ------------------------------------------------------------------------------
print("\n7. üöÄ Mod√©lisation avec Gradient Boosting (XGBoost)")

if 'X_train_scaled' not in locals() or 'X_test_scaled' not in locals() or 'y_train' not in locals() or 'y_test' not in locals():
    print("   ‚ùå Erreur : Les ensembles X_train_scaled, X_test_scaled, y_train ou y_test ne sont pas d√©finis.")
    print("   V√©rifie que la s√©paration des donn√©es et la normalisation ont √©t√© effectu√©es.")
else:
    # Importation des biblioth√®ques n√©cessaires
    from xgboost import XGBClassifier
    from sklearn.model_selection import cross_val_score
    from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, accuracy_score
    import joblib
    import numpy as np
    import pandas as pd

    # Initialisation du mod√®le avec des param√®tres optimis√©s pour les petits datasets
    model = XGBClassifier(
        random_state=42,
        eval_metric='logloss',
        use_label_encoder=False,
        n_estimators=100,
        max_depth=3,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8
    )

    # Validation crois√©e (5 folds) pour √©valuer la stabilit√© du mod√®le
    print("   Validation crois√©e en cours...")
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    print(f"   Scores de validation crois√©e (5 folds) : {cv_scores}")
    print(f"   Moyenne des scores : {np.mean(cv_scores):.2f} ¬± {np.std(cv_scores):.2f}")

    # Entra√Ænement du mod√®le sur l'ensemble d'entra√Ænement complet
    print("\n   Entra√Ænement du mod√®le sur l'ensemble complet...")
    model.fit(X_train_scaled, y_train)
    print("   ‚úÖ Mod√®le entra√Æn√© avec succ√®s.")

    # Importance des features (pour l'interpr√©tation clinique)
    feature_importance = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': model.feature_importances_
    }).sort_values(by='Importance', ascending=False)

    print("\n   üìä Top 5 des features les plus importantes :")
    print(feature_importance.head())

    # Pr√©dictions sur l'ensemble de test
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:, 1]

    # √âvaluation des performances
    print("\n   üìà Performance du mod√®le sur l'ensemble de test :")
    print(classification_report(y_test, y_pred, target_names=["Sain", "Malade"]))
    print(f"   Pr√©cision globale : {accuracy_score(y_test, y_pred):.2f}")
    print(f"   AUC-ROC : {roc_auc_score(y_test, y_proba):.2f}")
    print("   Matrice de confusion :")
    print(confusion_matrix(y_test, y_pred))

    # Sauvegarde du mod√®le (optionnel)
    joblib.dump(model, 'xgboost_heart_disease_model.pkl')
    print("\n   üíæ Mod√®le sauvegard√© sous 'xgboost_heart_disease_model.pkl'.")

# 8. OPTIMISATION DU MOD√àLE
# ------------------------------------------------------------------------------
print("\n8. üîß Optimisation du mod√®le pour am√©liorer la pr√©cision")

# --- A. Optimisation des hyperparam√®tres avec GridSearchCV ---
print("\n   A. Optimisation des hyperparam√®tres avec GridSearchCV")

from sklearn.model_selection import GridSearchCV

# D√©finir la grille de param√®tres
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

# Initialiser GridSearchCV
grid_search = GridSearchCV(
    estimator=XGBClassifier(random_state=42, eval_metric='logloss', use_label_encoder=False),
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',  # Optimisation pour maximiser l'AUC-ROC
    n_jobs=-1,          # Utilisation de tous les c≈ìurs disponibles
    verbose=1
)

# Ex√©cuter la recherche
print("   Recherche des meilleurs hyperparam√®tres en cours...")
grid_search.fit(X_train_scaled, y_train)

# Afficher les meilleurs param√®tres
print("\n   Meilleurs param√®tres trouv√©s :", grid_search.best_params_)

# Entra√Æner le mod√®le avec les meilleurs param√®tres
best_model = grid_search.best_estimator_
best_model.fit(X_train_scaled, y_train)

# √âvaluer le mod√®le optimis√©
y_pred_best = best_model.predict(X_test_scaled)
y_proba_best = best_model.predict_proba(X_test_scaled)[:, 1]

print("\n   Performance du mod√®le optimis√© :")
print(classification_report(y_test, y_pred_best, target_names=["Sain", "Malade"]))
print(f"   Pr√©cision globale : {accuracy_score(y_test, y_pred_best):.2f}")
print(f"   AUC-ROC : {roc_auc_score(y_test, y_proba_best):.2f}")
print("   Matrice de confusion :")
print(confusion_matrix(y_test, y_pred_best))

# --- B. R√©√©quilibrage des classes avec SMOTE ---
print("\n   B. R√©√©quilibrage des classes avec SMOTE")

from imblearn.over_sampling import SMOTE

# Appliquer SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

# Entra√Æner le mod√®le sur les donn√©es r√©√©quilibr√©es
model_res = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False,
    **grid_search.best_params_  # Utilisation des meilleurs param√®tres trouv√©s
)

model_res.fit(X_train_res, y_train_res)

# √âvaluer le mod√®le avec SMOTE
y_pred_res = model_res.predict(X_test_scaled)
y_proba_res = model_res.predict_proba(X_test_scaled)[:, 1]

print("\n   Performance du mod√®le avec SMOTE :")
print(classification_report(y_test, y_pred_res, target_names=["Sain", "Malade"]))
print(f"   Pr√©cision globale : {accuracy_score(y_test, y_pred_res):.2f}")
print(f"   AUC-ROC : {roc_auc_score(y_test, y_proba_res):.2f}")
print("   Matrice de confusion :")
print(confusion_matrix(y_test, y_pred_res))

# --- C. Interpr√©tation avec SHAP ---
print("\n   C. Interpr√©tation des r√©sultats avec SHAP")

import shap

# Initialiser l'explicateur SHAP pour le meilleur mod√®le
explainer = shap.TreeExplainer(best_model)

# Calculer les valeurs SHAP
shap_values = explainer.shap_values(X_test_scaled)

# Visualiser l'impact des features
plt.figure()
shap.summary_plot(shap_values, X_test_scaled, feature_names=X_test.columns, plot_type="bar")
plt.title("Impact des features sur les pr√©dictions (SHAP Values)")
plt.tight_layout()
plt.show()

plt.figure()
shap.summary_plot(shap_values, X_test_scaled, feature_names=X_test.columns)
plt.title("Contribution des features aux pr√©dictions (SHAP Values)")
plt.tight_layout()
plt.show()

# Sauvegarde du meilleur mod√®le
joblib.dump(best_model, 'xgboost_heart_disease_model_optimized.pkl')
print("\n   üíæ Meilleur mod√®le sauvegard√© sous 'xgboost_heart_disease_model_optimized.pkl'.")

# 9. √âVALUATION DES PERFORMANCES
# ------------------------------------------------------------------------------
print("\n8. üìä √âvaluation des performances du mod√®le")

# Importation des biblioth√®ques n√©cessaires
from sklearn.metrics import roc_curve, precision_recall_curve, average_precision_score, confusion_matrix, classification_report, accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# V√©rification que les mod√®les et les donn√©es sont bien d√©finis
models = {
    "Mod√®le de base": model,
    "Mod√®le optimis√©": best_model if 'best_model' in locals() else None,
    "Mod√®le avec SMOTE": model_res if 'model_res' in locals() else None
}

# Suppression des mod√®les non d√©finis
models = {k: v for k, v in models.items() if v is not None}

if not models:
    print("   ‚ùå Aucun mod√®le n'est d√©fini. V√©rifiez les √©tapes pr√©c√©dentes.")
else:
    # --- A. √âvaluation de chaque mod√®le ---
    results = []
    roc_data = {}
    pr_data = {}

    for name, model in models.items():
        print(f"\n   {name} :")

        # Pr√©dictions
        y_pred = model.predict(X_test_scaled)
        y_proba = model.predict_proba(X_test_scaled)[:, 1]

        # M√©triques
        acc = accuracy_score(y_test, y_pred)
        auc_roc = roc_auc_score(y_test, y_proba)
        report = classification_report(y_test, y_pred, target_names=["Sain", "Malade"], output_dict=True)
        cm = confusion_matrix(y_test, y_pred)
        precision, recall, _ = precision_recall_curve(y_test, y_proba)
        avg_precision = average_precision_score(y_test, y_proba)

        # Affichage des r√©sultats
        print(f"   - Accuracy : {acc*100:.2f}%")
        print(f"   - AUC-ROC : {auc_roc:.2f}")
        print(f"   - Average Precision : {avg_precision:.2f}")
        print("\n   - Rapport de classification :")
        print(classification_report(y_test, y_pred, target_names=["Sain", "Malade"]))

        # Matrice de confusion
        plt.figure(figsize=(6, 5))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                    xticklabels=["Sain", "Malade"], yticklabels=["Sain", "Malade"])
        plt.xlabel('Pr√©diction')
        plt.ylabel('R√©alit√©')
        plt.title(f'Matrice de Confusion ({name})')
        plt.tight_layout()
        plt.show()

        # Courbe ROC
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_data[name] = (fpr, tpr, auc_roc)

        # Courbe Precision-Recall
        pr_data[name] = (precision, recall, avg_precision)

        # Stockage des r√©sultats
        results.append({
            'Mod√®le': name,
            'Accuracy': acc,
            'AUC-ROC': auc_roc,
            'Average Precision': avg_precision,
            'Pr√©cision (Sain)': report['Sain']['precision'],
            'Pr√©cision (Malade)': report['Malade']['precision'],
            'Rappel (Sain)': report['Sain']['recall'],
            'Rappel (Malade)': report['Malade']['recall'],
            'F1-score (Sain)': report['Sain']['f1-score'],
            'F1-score (Malade)': report['Malade']['f1-score']
        })

    # --- B. Comparaison des courbes ROC ---
    plt.figure(figsize=(8, 6))
    for name, (fpr, tpr, auc) in roc_data.items():
        plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('Taux de Faux Positifs (FPR)')
    plt.ylabel('Taux de Vrais Positifs (TPR)')
    plt.title('Comparaison des Courbes ROC')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # --- C. Comparaison des courbes Precision-Recall ---
    plt.figure(figsize=(8, 6))
    for name, (precision, recall, avg_precision) in pr_data.items():
        plt.plot(recall, precision, label=f'{name} (AP = {avg_precision:.2f})')
    plt.xlabel('Rappel')
    plt.ylabel('Pr√©cision')
    plt.title('Comparaison des Courbes Precision-Recall')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # --- D. R√©sum√© des performances ---
    results_df = pd.DataFrame(results)
    print("\n   üìä R√©sum√© des performances :")
    display(results_df)

    # --- E. Analyse des erreurs du meilleur mod√®le ---
    best_model_name = results_df.loc[results_df['AUC-ROC'].idxmax(), 'Mod√®le']
    best_model_obj = models[best_model_name]

    print(f"\n   E. Analyse des erreurs du meilleur mod√®le ({best_model_name}) :")
    y_pred_best = best_model_obj.predict(X_test_scaled)
    cm_best = confusion_matrix(y_test, y_pred_best)

    # Identifier les faux n√©gatifs et faux positifs
    fp_indices = np.where((y_test == 0) & (y_pred_best == 1))[0]
    fn_indices = np.where((y_test == 1) & (y_pred_best == 0))[0]

    print(f"   - Nombre de faux positifs : {len(fp_indices)}")
    print(f"   - Nombre de faux n√©gatifs : {len(fn_indices)}")

    if len(fn_indices) > 0:
        print("\n   - Caract√©ristiques moyennes des faux n√©gatifs :")
        fn_features = pd.DataFrame(X_test_scaled[fn_indices], columns=X_test.columns)
        display(fn_features.describe().loc[['mean', 'std', 'min', 'max']])
    else:
        print("\n   - Aucun faux n√©gatif d√©tect√©.")

    # --- F. Analyse des pr√©dictions du meilleur mod√®le ---
    print("\n   F. Analyse des pr√©dictions du meilleur mod√®le :")
    y_proba_best = best_model_obj.predict_proba(X_test_scaled)[:, 1]
    predictions_df = pd.DataFrame({
        'Probabilit√©': y_proba_best,
        'Pr√©diction': y_pred_best,
        'R√©alit√©': y_test.values
    }).reset_index(drop=True)

    # Afficher les pr√©dictions avec les plus grandes erreurs
    predictions_df['Erreur'] = (predictions_df['Pr√©diction'] != predictions_df['R√©alit√©']).astype(int)
    error_cases = predictions_df[predictions_df['Erreur'] == 1].sort_values(by='Probabilit√©', ascending=False)
    print("\n   - Top 5 des erreurs de pr√©diction (faux positifs en haut) :")
    display(error_cases.head())

    # --- G. Visualisation des erreurs ---
    if len(fn_indices) > 0:
        print("\n   G. Visualisation des faux n√©gatifs :")
        fn_features = X_test_scaled[fn_indices]
        sns.pairplot(pd.DataFrame(fn_features, columns=X_test.columns), diag_kind='kde')
        plt.suptitle("Caract√©ristiques des faux n√©gatifs", y=1.02)
        plt.tight_layout()
        plt.show()

# --- FIN DU SCRIPT ---
print("\n--- FIN DE L'√âVALUATION DES PERFORMANCES ---")
```
